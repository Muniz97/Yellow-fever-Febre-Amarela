Yellow Fever Brazil 1994-2021 / Febre Amarela Brasil 1994-2021

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/Muniz97/Yellow-fever-Febre-Amarela/main/Datasets/fa_casoshumanos_1994-2021.csv'
```

```python
dados_raw = pd.read_csv(url, sep = ';', encoding="ISO-8859-1")
```

```python
# Display
pd.options.display.float_format = '{:.3f}'.format
```

```python
dados_raw.head(100)
```

```python
dados_raw.info()
```

Datas nulls and duplicates / Dados nulos e duplicados

```python
dados_raw.isnull().sum()
```

```python
dados_raw = dados_raw.dropna() # Por serem poucos dados nulos os mesmo foram removidos
```

```python
# Verifica atual volumetria
dados_raw.count()
```

```python
# Verifica se existe duplicados
dados_raw.drop_duplicates().count() # No datas duplicates / Sem dados duplicados
```

Convert features numerical / Convertendo features númericas

```python
# Separando features númericas
list_int = ['IDADE','SE_IS','MES_IS','ANO_IS'] # COD_UF_LPI and COD_MUN_LPI are objects
```

```python
# Convertendo
dados_raw['IDADE'] = dados_raw['IDADE'].str.replace(',', '.').astype(float)
```

```python
# int convert
dados_raw[list_int] = dados_raw[list_int].astype(int)
```

```python
dados_raw.info()
```

Datetime / Datas

```python
# Date
from datetime import datetime
```

```python
dados_raw.head(25)
```

```python
dados_raw.DT_IS
```

```python
dados_raw.DT_IS.unique()
```

```python
dados_raw['DT_IS'] = dados_raw['DT_IS'].apply(lambda x: datetime.strptime(x, "%d/%m/%Y")) # Valores fora do formato data foram removidos / brazilian date format
```

```python
dados_raw['DIA_IS'] = dados_raw['DT_IS'].dt.day # 'DIA' is day
```

String

```python
list_string = ['MACRORREG_LPI', 'UF_LPI', 'MUN_LPI', 'SEXO', 'MONITORAMENTO_IS' ,'OBITO'] # features are strings
```

```python
dados_raw[list_string] = dados_raw[list_string].applymap(lambda x: x.upper()) # applymap serve para quando se tem mais de um variavel
```

```python
dados_raw = dados_raw.drop(columns = dados_raw[['COD_UF_LPI','COD_MUN_LPI','DT_IS']], axis = 1)
dados_raw.head()
```
